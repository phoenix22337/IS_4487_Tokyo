```python
from sklearn.tree import plot_tree
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz # Import Decision Tree Classifier
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#advise invest
AV = pd.read_csv("https://raw.githubusercontent.com/jefftwebb/is_4487_base/main/Assignments/DataSets/adviseinvest.csv")

# Current customer data
CC = pd.read_csv("https://raw.githubusercontent.com/jefftwebb/is_4487_base/main/Assignments/DataSets/adviseinvest_customer_data.csv")
```

```python
CC.head()
```

```python
AV.head()
```

```python
AV = AV[(AV['income'] > 0) &
              (AV['num_accts'] < 10)]
AV['job'] = AV['job'].astype('category')
AV['product'] = AV['product'].astype('category')
AV['chk_acct'] = AV['chk_acct'].astype('category')
AV['sav_acct'] = AV['sav_acct'].astype('category')
AV = AV.dropna()
AV = AV.drop('product', axis=1)
```

```python
CC['job'] = CC['job'].astype('category')
CC['chk_acct'] = CC['chk_acct'].astype('category')
CC['sav_acct'] = CC['sav_acct'].astype('category')
```

```python
X = AV.drop(['answered'], axis=1)
y = AV['answered']

# initialize the tree
full_tree = DecisionTreeClassifier(criterion="entropy", max_depth = 5)

# Create Decision Tree Classifer
full_tree = full_tree.fit(X, y)
```

```python
X_new = CC.drop(['customer_id'], axis=1)
```

```python
pred = full_tree.predict_proba(X_new)[:, 1]
```

```python
CC['answer_prob'] = pred

CC = CC[['customer_id','answer_prob']]

CC
```

```python
sort_values = CC.sort_values(by=['answer_prob'], ascending=False)
sort_values[sort_values['answer_prob'] > .2]
```
